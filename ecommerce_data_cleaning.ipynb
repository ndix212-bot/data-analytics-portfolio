```python
# Convert all mixed up date strings into a single, clean standard format (YYYY-MM-DD)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

# Check our final column types to make sure everything looks pristine
df.info()
```

```python
# Show information about missing values and column data types
df.info()
```

```python
# Force Quantity to be numbers. Anything that isn't a number becomes an empty blank (NaN)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')

# Verify the change worked
df.info()
```

```python
# 1. Strip out any hidden dollar signs from the Price column
df['Price'] = df['Price'].astype(str).str.replace('$', '', regex=False)

# 2. Force the column to be a numeric float, turning bad data into blanks
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# 3. Verify that Price is now a float64
df.info()
```

```python
# Recalculate the entire Total column using our clean data
df['Total'] = df['Quantity'] * df['Price']

# Verify the changes and check how many non-null values we have now
df.info()
```

```python
df.to_csv('/content/cleaned_ecommerce_sales_data.csv', index=False)
```
